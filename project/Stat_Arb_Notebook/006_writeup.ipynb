# Leader--Follower Shock Propagation in Cryptocurrency Markets

## Overview

This project investigates whether large price shocks in one
cryptocurrency lead to predictable short‑term reactions in other
cryptocurrencies.\
The core hypothesis is that a **leader asset shock may temporarily
disrupt price relationships**, causing other assets to overreact and
later revert.

The research was implemented as a **multi‑stage quantitative research
pipeline**, progressing from raw data acquisition through signal
discovery, strategy construction, and backtesting.

------------------------------------------------------------------------

costs**.

# Key Empirical Observation

The analysis reveals a common pattern:

1. leader shock
2. followers react strongly
3. followers revert toward equilibrium

This temporary dislocation creates a small but repeatable trading opportunity.

However, the signal strength is modest and may require additional filtering or improved execution timing to become robust.

# Network Interpretation of Shock Propagation

Leader--follower relationships can be interpreted as a **directed graph of shock propagation** across assets.

In this framework:

-   each cryptocurrency is represented as a **node**
-   statistically significant leader--follower relationships are
    **directed edges**

Example relationships:

* BTC --> ETH
* ETH --> SOL
* BTC --> BNB

These relationships naturally form a **Directed Acyclic Graph (DAG)** when propagation is considered over short time windows.

Example propagation chain:

BTC --> ETH --> SOL

Interpretation:

1.  A large shock occurs in BTC
2.  ETH reacts shortly afterward
3.  SOL responds after ETH

In this view, shocks **diffuse through the market structure** rather
than occurring simultaneously across all assets.

------------------------------------------------------------------------

# Extending DAGs to Propagation Trees

Pairwise edges can be expanded into **propagation trees**.

Example structure:

        BTC
       /   \
     ETH   BNB
      |
     SOL

Here:

-   BTC acts as a primary shock source
-   ETH and BNB act as first‑order followers
-   SOL reacts as a second‑order follower

These structures describe **multi‑step propagation dynamics** where
shocks cascade across the market.

------------------------------------------------------------------------

# Implications for Strategy Design

Viewing the market as a shock propagation network suggests improvements:

1.  Pair filtering -- trade only statistically strong edges.
2.  Propagation depth -- detect second‑order reactions.
3.  Cluster detection -- identify groups of assets that respond together.
4.  Dynamic graph estimation -- update propagation networks over time.

------------------------------------------------------------------------

# Limitations

Several factors likely reduce strategy performance:

-   Daily data resolution may obscure intraday lead--lag relationships
-   The strategy currently trades all followers, including weak relationships
-   Market‑wide shocks may overwhelm the propagation effect

These limitations suggest opportunities for further research.

------------------------------------------------------------------------

# Future Research Directions

Potential extensions include:

-   Pair‑specific propagation modeling
-   Intraday shock detection
-   Shock breadth filtering
-   Graph‑based propagation models

Although the current implementation produces modest performance, the signal appears statistically meaningful and warrants deeper exploration.

strategies**.
