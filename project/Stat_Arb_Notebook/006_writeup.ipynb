# Leader–Follower Shock Propagation in Cryptocurrency Markets

## Overview

This project investigates whether large price shocks in one cryptocurrency lead to predictable short-term reactions in other cryptocurrencies.

The hypothesis is that a **leader asset shock temporarily disrupts cross-asset price relationships**, causing other assets to overreact and subsequently revert/recover.

To test this, a subset of 9 coins were chosen based on market capitalization, i.e. the mega fauna of the crypto market. Once selected, then research was conducted as a **multi-stage quantitative research pipeline**, progressing from data collection through signal discovery, strategy construction, and backtesting.

## Backtest Results

The strategy exhibits **strong performance under ideal execution conditions**, but struggles under further constraints.

In short, the observed signal is **real but fragile**, and highly dependent on execution details.

## Insights

Execution timing dominates this assessment. 

Performance is strongest at immediate execution (t+0), collapses at t+1, and partially recovers at longer delays. This non-monotonic pattern suggests the signal is driven by **short-lived microstructure effects**, rather than persistent inefficiencies.

Attempts to improve performance through simple selection criteria do not materially enhance results, indicating that the edge is **timing-driven rather than selection-driven**.

## Network Interpretation of Shock Propagation

Leader–follower relationships could also be interpreted as a **directed graph of shock propagation** across assets:

- each cryptocurrency represents a node  
- statistically observed relationships represent directed edges  

Example:

- BTC --> ETH  
- ETH --> SOL  
- BTC --> BNB  

These relationships form a **Directed Acyclic Graph (DAG)** over short horizons.

This suggests that shocks **propagate through the market structure**, rather than occurring simultaneously across assets.

## Assessment

The results indicate that:

- A **lead–lag effect exists** in cryptocurrency markets  
- The effect is **short-lived and highly timing-sensitive**  
- Performance is **strong under ideal conditions but fragile under realistic constraints**  
- The signal is **broad but weak**, rather than concentrated  

While statistically observable, the effect does not translate into a robust trading strategy under realistic assumptions.

The primary interpretation is that the signal reflects **market microstructure dynamics**, not a stable arbitrage opportunity.

## Recommended Improvements

Based on the findings, the following improvements are recommended:

### 1. Intraday Data

Daily resolution likely obscures the true dynamics of propagation.
- capture precise timing of reactions  
- align execution with signal window  
- identify sub-daily opportunities  

### 2. Volume Conditioning

Incorporate volume to differentiate signal quality:
- high-volume shocks → stronger information content  
- low-volume shocks → likely noise  

This may improve signal filtering and reduce false positives.

### 3. Graph-Based Modeling

Move beyond pairwise relationships:
- evaluate propagation paths  
- identify clusters of related assets  
- model multi-step transmission effects  

### 4. Selective Trade Construction

Avoid treating all events equally:
- rank opportunities  
- filter weaker signals  
- focus on higher-conviction subsets  

## Conclusion

This study identifies a **real but fragile lead–lag effect** in cryptocurrency markets.

The signal is strongest under immediate execution and low transaction costs, but deteriorates under more realistic conditions. This indicates that the observed effect is **short-lived, execution-sensitive, and unlikely to be directly exploitable without structural advantage**.

Rather than representing a deployable strategy, the results provide evidence of **microstructure-driven propagation effects**, forming a foundation for further research using higher-frequency data, improved filtering, and graph-based modeling approaches.
