**Author:** Wayne Kirk Schmidt

## Research Motivation

Cryptocurrency markets exhibit strong cross-asset relationships due to:

- shared liquidity pools  
- overlapping participant bases  
- correlated macro drivers  

However, price movements across assets are not always simultaneous.

Empirical observation suggests that **large movements in one asset propagate through the market with a short delay**, rather than occurring instantaneously.

This raises the possibility that **lead–lag relationships exist between major cryptocurrencies**. If such relationships are measurable and persistent, they may provide a basis for **statistical arbitrage strategies**.

## Research Hypothesis

Large price movements in a major cryptocurrency are followed by **statistically measurable reactions in other cryptocurrencies after a short time lag**.

Formally:

> Conditional on a large change in a return for asset A at time t,  
> the expected return of asset B at time t + k may be non-zero.

Testing this requires:

- synchronized historical price data  
- event-based analysis  
- cross-asset lag measurement  
- validation via backtesting

## Research Pipeline

The analysis is implemented as a staged notebook pipeline.

| stage | notebook | description |
|:-----|:---------|:------------|
| 000 | 000_overview.ipynb | structure, hypothesis, and design |
| 001 | notebooks/001_download.ipynb | data acquisition and event construction |
| 002 | notebooks/002_enrich.ipynb | derived datasets and lag features |
| 003 | notebooks/003_analysis.ipynb | shock structure and cross-asset relationships |
| 004 | notebooks/004_strategy.ipynb | signal construction |
| 005 | notebooks/005_backtest.ipynb | strategy evaluation |
| 006 | 006_writeup.ipynb | findings and conclusions |

Each stage produces artifacts consumed by subsequent stages, ensuring a **modular, reproducible, and auditable workflow**.

## Stage Definitions

### Stage 1 — Data Acquisition (001_download)

**Reads from:** external market data sources  
**Writes to:** `output/001_download/`  

**Artifacts:**
- `PRICES.pkl`  
- `<COIN>.event_panel.pkl`  

**Responsibilities:**

- define cryptocurrency universe  
- download daily price data  
- align time series across assets  
- compute daily returns  
- construct event panels  

### Stage 2 — Data Enrichment (002_enrich)

**Reads from:** `output/001_download/`  
**Writes to:** `output/002_enrich/`  

**Artifacts:**
- `returns_full.pkl`  
- `rolling_sigma.pkl`  
- `z_scores.pkl`  
- `event_bitmap.pkl`  

**Responsibilities:**

- expand lag windows  
- compute rolling volatility  
- normalize return signals  
- prepare datasets for analysis  

### Stage 3 — Analysis (003_analysis)

**Reads from:** `output/002_enrich/`  
**Writes to:** `output/003_analysis/`  

**Artifacts:**

- `leader_heatmap.png`  

**Responsibilities:**

- visualize lagged responses  
- identify leader–follower relationships  
- assess statistical structure of shocks  

### Stage 4 — Strategy Construction (004_strategy)

**Reads from:** `output/002_enrich/`, `output/003_analysis/`

**Artifacts:**

### Stage 5 — Backtest (005_backtest)

**Reads from:** `output/004_strategy/`

**Artifacts:**

### Stage 6 — Writeup (006_writeup)

**Reads from:** `output/005_backtest/`

**Responsibilities:**

- interpret results  
- evaluate hypothesis  
- assess robustness  
- propose future directions  

## Design Principles

The pipeline is designed around:

1. Determinism: Each stage produces well-defined artifacts with no hidden dependencies.
2. Modularity: Each stage can be executed independently and validated in isolation.
3. Reproducibility: All outputs are derived from persisted artifacts, enabling full reconstruction of results.
4. Separation of Concerns: are handled in distinct stages.
5. Local Exploration / Global Coordination

## In Scope

This project focuses on:

- major cryptocurrency assets  
- daily time resolution  
- short-term lead–lag relationships  

## Out of Scope

The project does not focus on:

- model intraday behavior  
- optimize execution  
- deployment of production trading systems  

# Expected Outcome

 While the goal of this project is to hopefully find a production-ready strategy,the primary focus is:

- identify whether a measurable lead–lag effect exists  
- characterize its structure  
- evaluate its robustness under realistic constraints
- determine under what conditions would a trading strategy be viable
- validate the methodology as a pre-cursor to Cirkus ( coordinated evaluation layer for assessment of trading strategies )

This document defines the structure, hypothesis, and execution plan for investigating lead–lag relationships in cryptocurrency markets.
