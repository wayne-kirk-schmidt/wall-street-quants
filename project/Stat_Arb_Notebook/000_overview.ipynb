## Research Motivation

Cryptocurrency markets exhibit strong cross-asset relationships due to shared liquidity pools, overlapping trader bases, and correlated macro drivers.

However, price movements across assets are not always simultaneous. In many cases, large movements in one asset appear to propagate through the market with a short delay.

This raises the possibility that **lead–lag relationships exist between major cryptocurrencies**.

If a leading asset moves sharply and other assets respond with a predictable delay, those patterns may provide the basis for **statistical arbitrage strategies**.

## Research Hypothesis

Large price movements in a major cryptocurrency are followed by statistically significant reactions in other major cryptocurrencies after a short time lag.

Formally:

Conditional on a large return in asset A at time t, the expected return of asset B at time t + k may be non-zero.

Detecting and validating such relationships requires:

- synchronized historical price data  
- event-based analysis  
- cross-asset lag measurements  
- statistical significance testing
