## Integer Indexing

```python
# Initialize variables

import numpy as np 

list1 = [1,1.1,-5.45,3.2]
arr1 = np.array(list1)

list2 = [[1,1.1,-5.45,3.2],[-0.5,0.33,10.12,-5]]
arr2 = np.array(list2)
```

```python
# 1d arrays are similar to python lists
print (arr1[0])
print (arr1[1:3])
print (arr1[-2:])
```

```python
# 2d array indexing
arr2
```

```python
arr2[0]
```

```python
arr2[0][1]
```

```python
arr2[0,1]
```

```python
arr2[0,:3]
```

```python
arr2[:,:3]
```

```python
list2[:,:3]
```

```python
arr2[:,[1,3,-1]]
```

```python
monty = [
    "spam", "spam", "spam", "spam",
    "bacon", "eggs", "sausage",
    "eggs", "cereal", "spam"
]

items, counts = np.unique(monty, return_counts=True)

hist = dict(zip(items, counts))
print(hist)
```
