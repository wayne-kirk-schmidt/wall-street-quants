# Numbers

```python
# 2 main numeric data types: integers and floats
num_int = 5
num_float = 2.5
```

```python
# use type() to get the data type
# notice python automatically infers the data type - no need to declare it separately
print(type(num_int))
print(type(num_float))
```

```python
x = 12
y = 8

add = x + y
subtract =  x - y
multiply = x * y 
divide = x / y 
modulus = x % y
exponent = x ** y
floor_div = x // y
```

```python
print (add)
print (subtract)
print (multiply)
print (divide)
print (modulus)
print (exponent)
print (floor_div)
```
